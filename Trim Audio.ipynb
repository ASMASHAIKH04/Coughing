```python
from pydub import AudioSegment

# Open an mp3 file
song = AudioSegment.from_file("/content/mild_coughing_8.mp3", format="mp3")

# start and end time
start_min = 0
start_sec = 17
end_min = 0
end_sec = 20

# pydub does things in milliseconds, so convert time
start = ((start_min*60)+start_sec)*1000
end = ((end_min*60)+end_sec)*1000

# song clip of 10 seconds from starting
first_10_seconds = song[start: end]

# save file
first_10_seconds.export("/content/mild_coughing_23.mp3", format="mp3")
print("New Audio file is created and saved")
```

```python
# Install pydub if not already installed
!pip install pydub

from google.colab import files
from pydub import AudioSegment

# Assuming 'files' is defined after uploading
uploaded = files.upload()

audio_file = 'normal_speech'
extension = list(uploaded.keys())[0].split('.')[-1]
full_audio_file = f'{audio_file}.{extension}'

# Load the audio file
audio = AudioSegment.from_file(full_audio_file)

# Duration of each clip in milliseconds
clip_duration_ms = 2000

# Generate and save the clips
for i in range(150):
    start_time = i * clip_duration_ms
    end_time = start_time + clip_duration_ms
    clip = audio[start_time:end_time]
    clip_file_name = f'{audio_file}_clip_{i+1}_02.{extension}'
    clip.export(clip_file_name, format=extension)
    print(f'Clip {i+1} saved as {clip_file_name}')

# Download each clip
print("Downloading clips...")
for i in range(150):
    clip_file_name = f'{audio_file}_clip_{i+1}_02.{extension}'
    files.download(clip_file_name)
```
